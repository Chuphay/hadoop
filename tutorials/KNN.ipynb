# KNN Tutorial

### Introduction

K-Nearest Neighbors is arguably the easiest machine learning algorithm to both understand and to program. Let us look at a few examples to help us get a feel for how this woks on an intuitive level.

Suppose, you are a real estate agent and you need to give an estimate for how much a house will sell for. A natural and intuitive place to start would be to see how much houses in the surrounding area sold for. 

Or suppose you were a doctor checking to see if someone had a chance of developing a heart disease. A reasonable way to proceed would be to find other patients with the same blood pressure, age, etc. and see if they developed a heart disease.

These two examples, I hope, revealed what it means to look at your nearest neighbors to be able to predict new outcomes. To make a prediction, we look at historical data and find the nearest neighbor. 

For the mathematically minded, you might be wondering how we decide what is "near". It turns out that there are many ways to define "near". But for the data that I will provide, the simplest way to define nearness is simply using Euclidean distance, i.e., $\sqrt{x^2 + y^2}$. 

Now that we understand what a qualifies as nearest neighbor, the question immediately is what is this K? Well, if you are guaranteed noiseless data, then simply picking the nearest neighbor is probably a good idea. However, if there is noise in the data, as there almost always is, then it would be better to pick several nearest neighbors and letting the majority win to tell us how to classify the data point in question.



### KNN on Small Data Sets

Before moving onto a big data set, we should test our understanding of the algorithm with a small data set. 

If you are new to python, the following lines set up python so that it can plot data, and we also import numpy which provides some usefull mathematical tools.

````python
%pylab inline
import numpy as np
````

After setting up python, I made a script to generate some artificial data. Artificial data is especially useful when you are testing an algorithm because you have a really good idea fo what is going on.

````python
def make_knn_data(num):
    x = np.random.uniform(0,10,size = num)
    y = np.random.uniform(0,10,size = num)
    target = np.array(['z' for i in range(num)])
    for i in range(num):
        if (x[i]**2 + y[i]**2 < 80):
            if(y[i] - x[i] < 3):
                target[i] = 'a'
            else:
                target[i] = 'b'
        else:
            target[i] = 'c'
        
        if(np.random.uniform()<0.1):
            target[i] = np.random.choice(['a','b','c'],1)[0]
    return (np.array([x,y]).T, target)      
````

After making this function to generate the data, I also made a small function to plot the data so that we can see what is going on.

````python
def plot_data(data, target):
    for i in range(len(target)):
        if(target[i] == 'a'):
            scatter(data[i][0],data[i][1], color = 'red')
        elif(target[i] == 'b'):
            scatter(data[i][0],data[i][1], color = 'blue')
        else:
            scatter(data[i][0],data[i][1], color='black')
                
seed(1234)       
data, target = make_knn_data(500)  
plot_data(data, target)
````

Up above is the data that we will be looking at. First of all, remember that this is artificial data and does not really have any meaning, but if it helps you, we can think of this as the prices of houses: red is cheap, black is average, and blue is expensive.

We also want to notice that the data is noisy. This means that we will definately want to choose more than one neighbor when we are making a classification.

Below I made a function to calculate Euclidean distance and then I made the function that actually gives me the K-Nearest Neighbors.

Let's walk through the KNN algorithm to see exactly what is going on. First the arguments to the function are the target coordinates for which we want to make a prediction, denoted targetX and targetY, plus the training data, and the target data, and finally k is the number of nearest neighbors.

We then initialize an array of distances. We then iterate over each data point in the data and calculate the distance from that data point to the target data point. We add this distance to the originally empty array distance.

Finally, we sort the data according to distance and return the k smallest distances with their appropriate classification. (If you have never seen zip before, you can probably look at how it works by googling it).

````python
def distance(targetX, targetY, x, y):
    return ((targetX - x)**2 + (targetY - y)**2)**0.5

def KNN(targetX, targetY, data, target, k = 3):
    distances = []
    for i in range(len(target)):
        distances.append(distance(targetX, targetY, data[i][0], data[i][1]))
    out = sorted(zip(distances, target))
    return out[:k]
    
        
KNN(6,7,data, target, 10)    
````

Looking at the nearest neighbors to the point (6,7) we can see that the closest neighbor is 'b' but the majority of the data (i.e., 60% of the data) is classified as 'c'. This is why we do not want to choose only the closest neighbor if there is any noise in the data, but rather take k nearest neighbors so that we can average over the answers to choose the best answer.

### Map Reduce

Now we are ready to code up k-Nearest Neighbors as a map reduce job. Two possiblities are likely to occur in a real life situation. One is that there are many files that have an assortment of data similar to the data preseented here. The other is that at some point someone has ordered the data, and sorted the data in some way for ease of access. We will however only examine the former.

We assume that we have several (or many) files that each contain data of the form above. So that a typical file will have a listing of data like:

```
 1.91519453  8.83951165 c
 6.22108771  7.41360534 c
 4.37727739  5.15710799 a
 7.85358584  1.35251844 a
 ...
 ...
 ...
``` 

It is extremely trivial to program up a map reduce job. Each mapper will be responsible for one file, or if the files are particularly big, then each mapper will only be in charge of a piece of each file. The person that runs the program should supply the program with the point that we want to find the nearest neighbors and how many nearest neighbors to find. The reducer does very little, it simply gathers up all the nearest neighbors that the various mappers generated and then chooses the k neighbors with the smallest distance. 

A small example may help us here. Let's use the tools we have already made to see how this would work. We will first generate 5 data sets

````python
seed(1234)       
data1, target1 = make_knn_data(500) 
data2, target2 = make_knn_data(500)
data3, target3 = make_knn_data(500)
data4, target4 = make_knn_data(500)
data5, target5 = make_knn_data(500)
````

After collecting the data, our mapper will simply consist of finding the k nearest neighbors on each data set. Let's choose k = 2 for this example, and still look at the point (6,7):

````python
map1 = KNN(6, 7, data1, target1, 2)   
print(map1[0][0], map1[0][1])
print(map1[1][0], map1[1][1])
map2 = KNN(6, 7, data2, target2, 2) 
print(map2[0][0], map2[0][1])
print(map2[1][0], map2[1][1])
map3 = KNN(6, 7, data3, target3, 2) 
print(map3[0][0], map3[0][1])
print(map3[1][0], map3[1][1])
map4 = KNN(6, 7, data4, target4, 2) 
print(map4[0][0], map4[0][1])
print(map4[1][0], map4[1][1])
map5 = KNN(6, 7, data5, target5, 2) 
print(map5[0][0], map5[0][1])
print(map5[1][0], map5[1][1])
````

The reducer simply takes this data and then selects the k nearest neighbors. We want to notice two things here. First of all k does not have to be the same k that occured with the mapper. We can choose any new k that is less than the old k multiplied by the number of mappers that you have. Here we can see that given an original k of 2, and 5 mappers, we can choose a final k as big as 10.

An even more import observation that will make our lives much simpler is to have a little understanding of how Hadoop works. After Hadoop finishes all the map jobs, there is a step that is called the "sort/shuffle phase" what this means is that the output from the mappers will be sorted based on the first number, also known as the "key" in Hadoop. Therefore, the input to the reducer is not as we see above, but is rather like the following:

````python
print(map3[0][0], map3[0][1])
print(map4[0][0], map4[0][1])
print(map4[1][0], map4[1][1])
print(map2[0][0], map2[0][1])
print(map1[0][0], map1[0][1])
print(map2[1][0], map2[1][1])
print(map1[1][0], map1[1][1])
print(map3[1][0], map3[1][1])
print(map5[0][0], map5[0][1])
print(map5[1][0], map5[1][1])
````

Because of the "sort/shuffle phase", programming up the reducer is extremely simple. We simple take the first k elements that come and ignore the rest.
